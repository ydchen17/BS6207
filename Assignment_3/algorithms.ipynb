## Max pooling

```
mp = torch.nn.MaxPool2d(kernel_size=2, stride=1, padding=0, dilation=1, return_indices=False, ceil_mode=False)
torch_out = mp(input)[0]
torch_out, torch_out.shape
```

```
def maxpool2d(input, kernal_size, stride):
    minibatch = input.shape[0]
    channel_size = input.shape[1]
    img_size = input.shape[2]
    output = torch.zeros(minibatch, channel_size, int((img_size - kernal_size) / stride) + 1, int((img_size - kernal_size) / stride) + 1)
    for pic_i in range(minibatch):
        for channel_i in range(channel_size):
            for img_i in range(output.shape[2]):
                for img_j in range(output.shape[2]):
                    list = []
                    for kernal_i in range(kernal_size):         
                        for kernal_j in range(kernal_size): 
                            list.append(input[pic_i][channel_i][img_i*stride+kernal_i][img_j*stride+kernal_j])
                    output[pic_i][channel_i][img_i][img_j] = max(list)
    return output

my_out = maxpool2d(input, kernal_size=2, stride=1)[0]
```

```
torch.sum(my_out - torch_out) < 1e-6
```

## Average pooling

```
ap = torch.nn.AvgPool2d(kernel_size=2, stride=1, padding=0, ceil_mode=False, count_include_pad=True, divisor_override=None)
torch_out = ap(input)[0]
torch_out, torch_out.shape
```

```
def avgpool2d(input, kernal_size, stride):
    minibatch = input.shape[0]
    channel_size = input.shape[1]
    img_size = input.shape[2]
    output = torch.zeros(minibatch, channel_size, int((img_size - kernal_size) / stride) + 1, int((img_size - kernal_size) / stride) + 1)
    for pic_i in range(minibatch):
        for channel_i in range(channel_size):
            for img_i in range(output.shape[2]):
                for img_j in range(output.shape[2]):
                    list = []
                    for kernal_i in range(kernal_size):         
                        for kernal_j in range(kernal_size): 
                            list.append(input[pic_i][channel_i][img_i*stride+kernal_i][img_j*stride+kernal_j])
                    output[pic_i][channel_i][img_i][img_j] = sum(list)/len(list)
    return output

my_out = avgpool2d(input, kernal_size=2, stride=1)
```

## Conv2D (dilation = 1)

```
torch_conv2d = torch.nn.Conv2d(in_channels=3, out_channels=6, kernel_size=3, stride=1, padding=0, dilation=1, groups=1, bias=True, padding_mode='zeros')
torch_out = torch_conv2d(input)[0]
torch_out, torch_out.shape
```

```
def conv2d(input, out_channels, kernel_size, stride, dilation, weight_from_torch, bias_from_torch):
    minibatch = input.shape[0]
    channel_size = input.shape[1]
    img_size = input.shape[2]
    dilation_kernel_size = kernel_size + (kernel_size-1)*(dilation-1)
    output = torch.zeros(minibatch, out_channels, int((img_size - dilation_kernel_size) / stride) + 1, 
                         int((img_size - dilation_kernel_size) / stride) + 1)
    for pic_i in range(minibatch):
        for out_channel_i in range(out_channels):
            for channel_i in range(channel_size):
                for img_i in range(output.shape[2]):
                    for img_j in range(output.shape[2]):
                        sum_list = []
                        for kernal_i in range(kernel_size):         
                            for kernal_j in range(kernel_size): 
                                sum_list.append((input[pic_i][channel_i][img_i*stride+kernal_i*dilation][img_j*stride+kernal_j*dilation])*
                                (weight_from_torch[out_channel_i][channel_i][kernal_i][kernal_j]))
                        output[pic_i][out_channel_i][img_i][img_j] = output[pic_i][out_channel_i][img_i][img_j] + sum(sum_list)
            output[pic_i][out_channel_i] = output[pic_i][out_channel_i]+bias_from_torch[out_channel_i]
    return output
my_out = conv2d(input, out_channels=6, kernel_size=3, stride=1, dilation=1, weight_from_torch = torch_conv2d.weight, bias_from_torch=torch_conv2d.bias)[0]
my_out, my_out.shape
```

```
torch.sum(my_out - torch_out) < 1e-6
```

## ConvTranspose2d

```
torch_ct = torch.nn.ConvTranspose2d(in_channels=3, out_channels=4, kernel_size=3, stride=1, padding=0, output_padding=0, groups=1, bias=True, dilation=1, padding_mode='zeros')
torch_out = torch_ct(input)
torch_out
```

```
def convT(x, in_channels, out_channels, kernel_size, stride, weight, bias, dilation):

    x_batch = x.size(0)
    x_channel = x.size(1)
    x_height = x.size(2)
    x_width = x.size(3)
    
    kernel_height = kernel_size[0]
    kernel_width = kernel_size[1]

    y_height = (x_height + kernel_height) - 1
    y_width = (x_width + kernel_width) - 1
    y_channels = out_channels
    y = torch.zeros((x_batch, y_channels, y_height, y_width))
    
    for a in range(x_batch):
      for b in range(y_channels):
        for c in range(x_channel):
          for i in range(x_height):
            for j in range(x_width):
              y[a, b, i:i + kernel_height, j:j + kernel_width] += x[a, c, i, j] * weight[c][b]
    
    for i in range(y_channels):
      y[0][i] += bias[i]

    return y

my_out = convT(input, in_channels = 3, out_channels = 4, kernel_size = (3, 3), stride = 1, weight = torch_ct.weight, bias = torch_ct.bias, dilation = 1)
my_out
```

```
torch.sum(my_out - torch_out) < 1e-6
```

## roi_pool

```
# https://zhuanlan.zhihu.com/p/104361606
boxes = torch.Tensor([[0, 0, 0, 15, 15]])
torch_out = torchvision.ops.roi_pool(input, boxes, (4, 4))
torch_out
```

```
def my_roi_pool(X, boxes, output_size):
    
    roi_start_height = int(boxes[0][1])
    roi_start_width = int(boxes[0][2])
    roi_end_height = int(boxes[0][3])
    roi_end_width = int(boxes[0][4])
    
    roi_width = roi_end_width - roi_start_width + 1
    roi_height = roi_end_height - roi_start_height + 1

    bin_width = roi_width / output_size[0]
    bin_height = roi_height / output_size[1]
    
    result = []
    for i in range(X.shape[1]):
      tmp = []
      for j in range(output_size[0]):
        r = []
        for k in range(output_size[1]):
          temp = X[..., int(roi_start_height + bin_height * j):math.ceil(roi_start_height + bin_height * (j + 1)), 
                   int(roi_start_width + bin_width * k):math.ceil(roi_start_width + bin_width * (k + 1))][0][i]
          r.append(temp.max())
        tmp.append(r)
      result.append(tmp)

    result = torch.Tensor(result)
    result = result.unsqueeze(0)
    return result
  
my_out = my_roi_pool(input, boxes, (4, 4))
my_out
```

```
torch.sum(my_out - torch_out) < 1e-6
```

```
running_mean = torch.Tensor([input[0][i].mean() for i in range(input.shape[1])])
running_var = torch.Tensor([input[0][i].var() for i in range(input.shape[1])])

def batch_norm(input,mean,var, momentum, eps):
    output = torch.zeros(input.shape)
    for i in range(input.shape[1]):
        output[0][i] = (input[0][i] - mean[i])/(np.sqrt(var[i])-eps)
    return output
my_out = batch_norm(input, running_mean, running_var, momentum=0.1, eps=1e-05)
my_out
```

## cross_entropy

```
input = torch.randn([1, 3, 32, 32])
target = torch.randint(3, (1, 32, 32), dtype=torch.int64)
```

```
torch_out = torch.nn.functional.cross_entropy(input, target, weight=None, size_average=None, ignore_index=-100, reduce=None, reduction='mean')
torch_out
```

```
def softmax(x):
    res = torch.zeros(x.shape)
    for i in range(len(x)):
        exp = torch.exp(x[i])
        sum = torch.sum(exp, dim=0)
        softmax = exp/sum
        res[i] = -torch.log(softmax)
    return res
    
def nn_loss(input, target):
    loss_list = torch.zeros([len(input)])
    for k in range(len(input)):
        loss = 0
        for i in range(target[k].shape[1]):
            for j in range(target.shape[2]):
                index = target[k, i, j]
                num = input[k, index, i ,j]
                loss -= num
        loss_list[k] = -loss / target.shape[1] / target.shape[2]
    return loss_list

def cross_entropy(input, target):
    log_softmax = softmax(input)
    loss = nn_loss(log_softmax, target)
    return loss.mean()

my_out = cross_entropy(input, target)
my_out
```

```
torch.sum(my_out - torch_out) < .01
```
